### Plot of Age vs Salary

*We want to make a plot of the cols age and salary, so we have to take the index of those 2 cols (the index of col is the name of that col unlike the index of a row which is numerical).*

            --------------------------------------------------------------------------------------
    Text(0.5, 1.0, 'Salary of Employees based on their Age'): This is the representation of a text object in Matplotlib.

Breakdown

-Text: This indicates that the object is a piece of text.

-(0.5, 1.0): These are the coordinates where the text is placed.
        0.5 represents the x-coordinate.
        1.0 represents the y-coordinate.
        In this context, these coordinates are normalized between 0 and 1, where (0,0) is the bottom-left of the plot and (1,1) is the top-right. Thus, (0.5, 1.0) places the text centered horizontally and at the top of the plot.

-'Salary of Employees based on their Age': This is the actual text content, which is the title of your graph.

```python
#Creating a line plot

plt.plot(file['Age'],file['Salary'],linestyle='dashed',marker='*',color='purple',markerfacecolor='yellow',markeredgecolor='black')  #plot(x-axis,y-axis)
#linestyle is for the type of line you want - dashed, dashdot etc. Default is the continuous line.
#marker marks the points of values and we can use symbols like x,o etc for it.

#plt.figure(facecolor='lightblue')
plt.xlabel('Age')   #to label the x-axis representing age
plt.ylabel('Salary')  #to label the y-axis representing salary
plt.title('Salary of Employees based on their Age')  #Title of the plot
```

### Graph depicting salary w.r.t. name

```python
#Creating a Bar Graph
plt.bar(file['Name'],file['Salary'],color='black')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary of each employee')
```

### Frequency of Age for the Employees

Frequency of the values - how many times a value is repeating

```python
#Histogram
plt.hist(file['Age'],color='lightblue',range=(10,50))
plt.xlabel('Age')
plt.ylabel('Count of Age')
plt.title('Frequency of age of all employees')
plt.savefig('histo.png')   #to save the figure
plt.show() #to display chart but latest versions don't need it
```

```python
#creating scatter Plot - to find the relation bw x and y axis
colors=['red','green','blue','purple']
plt.scatter(file['Age'],file['Salary'],c=colors)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot')
plt.grid(True)   #to form grid lines
```

## Advanced charts
### Pie chart of % count of age

```python
age_count=file['Age'].value_counts()  #will return count of each value within column
plt.pie(age_count,labels=age_count.index,autopct='%1.1f%%')    #index is for printing name of the column,autopct for printing percentage
#1.1f will print 1 digit after . ; 1.2f for 2 digits ; d for integer
plt.title('Count of ages')
```

### Box plot for distribution of salaries

- The box depicts IQR (Inter Quartile Range). IQR is a range in which almost 50% of values are there.
- The orange line inside the figure is Median
- IQR has 3 parts: Q1(depicting lowest value of IQR),Q2(median value),Q3(highest value)
- Q2' is mean
- outliers - o :values very large from other values in the data

```python
plt.boxplot(file['Salary'])
plt.title('Dispersion of salary')
plt.ylabel('Salary')
```

### Grid of the charts

fig - for main chart
axs - for sub chart

**fig, axs = plt.subplots(2,2,figsize=(10,10))**
- plt.subplots(2, 2): This function creates a figure and a 2x2 grid of subplots. The first argument 2 specifies the number of rows and 2nd one the columns.
- figsize=(10, 10): This argument sets the size of the figure. figsize is a tuple where the first value (10) represents the width of the figure in inches, and 2nd one (10) represents the height.
- fig, axs: fig is the Figure object, which is the entire figure or collection of subplots.
    axs is an array of Axes objects, which are the individual subplots. Since we have a 2x2 grid, axs will be a 2x2 array (nested list) of Axes objects.

```python
#STEP 1 : Declaration to create sub-plot
fig, axs = plt.subplots(2,2,figsize=(10,10))

#line chart
axs[0,0].plot(file['Age'],file['Salary'])
axs[0,0].set_title('Line Plot')    

#bar graph
axs[0,1].bar(file['Name'],file['Salary'])
axs[0,1].set_title('Bar Graph')  

#histogram
axs[1,0].hist(file['Age'])
axs[1,0].set_title('Histogram')  

#scatter plot
axs[1,1].scatter(file['Age'],file['Salary'])
axs[1,1].set_title('Scatter Plot')  

plt.tight_layout()   #to avoid overlapping of the figures
```
