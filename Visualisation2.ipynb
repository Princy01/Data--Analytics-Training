```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

- np.random.choice is a function from the NumPy library that generates a random sample from a given array.
    ['A', 'B', 'C', 'D'] is the array from which the random choices will be made.
    size=100 specifies that 100 random elements should be chosen from the array.
    The result is a NumPy array of 100 elements, where each element is randomly chosen from 'A', 'B', 'C',       or 'D'.

- np.random.rand is a function from the NumPy library that generates an array of random numbers.
    The numbers are drawn from a uniform distribution over the interval [0, 1).
    The argument 100 specifies the size of the array, meaning it will contain 100 random numbers.

    np.random.rand(100) * 100: Each random number generated by np.random.rand(100) is multiplied by 100.
    This scales the random numbers to be in the range [0, 100).


- pd.date_range(start='1/1/2022', periods=100) will generate dates from 2022-01-01 to 2022-04-10 (100 days starting from January 1, 2022).

```python
np.random.seed(0)   #this will ensure that random values are generated from 0
data={ 'Category':np.random.choice(['A','B','C','D'],size=100),
      'Value1':np.random.rand(100)*100,
      'Value2':np.random.rand(100)*50,
      'Date':pd.date_range(start='1/1/2022',periods=100)}

file=pd.DataFrame(data)
```

### Scatterplot of Seaborn

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,6))
sns.scatterplot(x='Value1',y='Value2',data=file)
plt.title('Value 1 vs Value 2',fontsize=18)
plt.xlabel('Value1',fontsize=14)
plt.ylabel('Value1',fontsize=18)
```

### Barplot

```python
sns.barplot(x='Category',y='Value1',data=file)
plt.title('Value1 by Category')
```

### Value1 over a period of time

```python
sns.lineplot(x='Date',y='Value1',data=file)
plt.xticks(rotation=90)
plt.title('value 1 with time')
```

### Histogram of sns

```python
sns.histplot(file['Value1'],bins=10,kde=True)
plt.title('Distribution of Value1')
```

### Boxplot

```python
sns.boxplot(x='Category',y='Value1',data=file)
plt.title('Value2 by category')
```

### violin plot

```python
sns.violinplot(x='Category',y='Value2',data=file)
plt.title('Value2 by Category')
```

### Pairplot

```python
sns.pairplot(file)
```

### Heatmap

heatmap defines co-linearity(independent variables being dependent on each other)

```python
sns.heatmap(file[['Value1','Value2']].corr(),annot=True)
plt.title('Correlation Matrix')
```

### Joint plot

```python
sns.jointplot(x='Value1',y='Value2',data=file,kind='reg')
```

### Swarm plot

```python
sns.swarmplot(x='Category',y='Value1',data=file)
print('Value by category')
```
