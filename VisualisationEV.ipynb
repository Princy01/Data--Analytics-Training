# IEA Global EV Data 2024

```python
# Importing necessary libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Data Reading

```python
data=pd.read_csv('IEA_EVdata.csv',low_memory=False)     #Reading the dataset
file=pd.DataFrame(data)   #The dataset is loaded into a DataFrame named file.
print(file)
```

### Data Inspection

```python
# Displaying the first few rows of the dataset
print("First few values: \n")
file.head()
```

```python
# Displaying the last few rows of the dataset
print("Last few values: \n")
file.tail()
```

```python
# Displaying the information about the dataset
file.info()
```

```python
# Counting the number of null values in each column

null_Val=file.isnull().sum() 
print("No. of null values: \n",null_Val)
```

## Data Visualization

### Region vs Category Plot

```python
plt.figure(figsize=(10,6))
plt.plot(data['region'], data['category'], marker='o', color='brown')
plt.xticks(rotation=90)
plt.title('Region with category')
plt.xlabel('Region')
plt.ylabel('Category')
```

### Value by Year Bar Plot

```python
sns.barplot(x='year',y='value',data=file)
plt.title('Value by year')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Value')
```

### Region Frequency Histogram

```python
plt.figure(figsize=(10,6))
sns.histplot(file['region'],bins=10,kde=True,color='black')
plt.title('Frequency of region')
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Count of Region')
```

### Region vs Parameter Scatter Plot

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
sns.scatterplot(x='region',y='parameter',data=file)
plt.title('Region with parameter',fontsize=18)
plt.xlabel('region',fontsize=14)
plt.ylabel('parameter',fontsize=18)
```

### Region Count Pie Chart

```python
sliced_data=file[:1000]
plt.figure(figsize=(10,6))
region_count=sliced_data['region'].value_counts()  #will return count of each value within column
plt.xticks(rotation=90)
plt.pie(region_count,labels=region_count.index,autopct='%1.1f%%')
plt.title('Count of region')
```

### Heatmap of Correlation Matrix

```python
sns.heatmap(file[['year','value']].corr(),annot=True)
plt.title('Heatmap of Year and Values')
```

### Pairplot

```python
sns.pairplot(file[:1000])
print('Pairplot of first 1000 entries:')
```

### Category vs Value Swarm Plot

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(x='category', y='value', data=file[:100])
plt.title('Category vs Value')
```
