# Data Analytics of IPL Matches 

```python
#Importing necessary libraries

import matplotlib.pyplot as plt #for visualisation of complex data
import seaborn as sns #for visualisation of 
import numpy as np  # linear algebra
import pandas as pd # data processing
```

```python
data=pd.read_csv('matches.csv',low_memory=False)     #Reading data from csv
df=pd.DataFrame(data)                                #The dataset is loaded into a DataFrame named df.
df
```

## Data Visualization

### Heatmap of Correlation Matrix

```python
sns.heatmap(df[['id','season','dl_applied','win_by_runs','win_by_wickets']].corr(),annot=True)
plt.title('\nCorrelation Matrix\n')
```

**This shows the columns 'id' and 'season' show the maximum correlation.**

### ID vs Season Plot

```python
plt.figure(figsize=(10,6))
plt.plot(data['id'], data['season'],linestyle='dashed',marker='o',color='purple',markerfacecolor='purple',markeredgecolor='black')
plt.xticks(rotation=90)
plt.title('ID with Season\n')
plt.xlabel('ID')
plt.ylabel('Season\n')
```

**This visualization underscores the strong correlation between id and season, highlighting the orderly structure of the IPL dataset.** 

### City Frequency Histplot

```python
plt.figure(figsize=(10,6))
sns.histplot(df['city'],color='brown',bins=10,kde=True)
plt.xticks(rotation=90)
plt.title('City Frequency\n')
```

**This shows Mumbai was chosen most of the times as the city to play IPL matches.**

### Season with Venue Scatter Plot

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)
sns.scatterplot(x='season',y='venue',color='red',data=df)
plt.title('Season with Venue\n',fontsize=18)
plt.xlabel('\nSeason',fontsize=18)
plt.ylabel('Venue\n',fontsize=18)
```

**This scatter plot effectively demonstrates the distribution of IPL matches across different seasons and venues.**

### Region Count Pie Chart

```python
plt.figure(figsize=(20,20))
region_count=df['winner'].value_counts()  #will return count of each value within column
plt.xticks(rotation=90)
color=['lightblue','pink','lightgrey','lightgreen','yellow']
plt.pie(region_count,labels=region_count.index,autopct='%1.1f%%', textprops={'fontsize': 16},colors=color)
plt.title('\nCount of region',fontsize=29)
```

**This pie chart shows the team "Mumbai Indians" has won the most matches.**

### Barplot for winner and Wins By Runs

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation=90,fontsize=25)
plt.yticks(fontsize=25)
sns.barplot(x='winner',y='win_by_runs', hue='winner', data=df, palette='Set3')
plt.xlabel('\nWinner', fontsize=25)
plt.ylabel('Win by Runs\n', fontsize=25)
plt.title('\nDistribution of Win by Runs Across Winners\n',fontsize=40)
```

**This barplot demonstrates that Chennai Super Kings had the most wins by runs while Kochi Tuskers Kerala had the least.**

### Violin plot for Winner and Win By Wickets

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation=90,fontsize=25)
plt.yticks(fontsize=25)
sns.violinplot(x='winner',y='win_by_wickets',data=df)
plt.xlabel('\nWinner', fontsize=25)
plt.ylabel('Win by Wickets\n', fontsize=25)
plt.title('\nDistribution of Win by wickets across Winners\n',fontsize=40)
```

**The violin plot displayed illustrates the distribution of wins by wickets for different IPL match winners.**
