```python
#Importing necessary libraries

import matplotlib.pyplot as plt #for visualisation of complex data
import seaborn as sns #for visualisation of 
import numpy as np  # linear algebra
import pandas as pd # data processing
from sklearn.linear_model import LinearRegression #for algo of linear regression
from sklearn.metrics import mean_squared_error,r2_score #for evaluation of prediction model
```

```python
#Step-1: Create random data/read the file

#random feature(input) creation
X=np.random.rand(100,1)*10

#random label(output) creation
Y=2.5*X+np.random.rand(100,1)*2
```

```python
#Converting lists random into dataframe
df_X=pd.DataFrame(X,columns=['Feature'])
df_Y=pd.DataFrame(Y,columns=['Target'])
```

```python
#Combine both X and Y dataframe
data=pd.concat([df_X,df_Y],axis=1)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

### Creating scatterplot of random generated data

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,6))
#plt.xticks(rotation=90)
sns.scatterplot(x='Feature',y='Target',color='red',data=data)
plt.title('Scatter Plot of Feature vs Target\n',fontsize=18)
plt.xlabel('\nFeature',fontsize=18)
plt.ylabel('Target\n',fontsize=18)
```

## Training the model
### a) Implementation of Algorithm

```python
model=LinearRegression()
```

### b) Execute this trained algo on your dataset (fitting of algorithm)

```python
model.fit(X,Y)
```

### c) Letting it to predict the values

```python
 #insert a variable  in this parameter to which you want to predict the value
y_pred=model.predict(X) 
print(y_pred)
```

### Scatter plot with regression line

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,6))
#plt.xticks(rotation=90)
sns.scatterplot(x='Feature',y='Target',color='brown',data=data,label="Data")
plt.plot(X,y_pred,label='Regression Line')
plt.title('Scatter Plot with Regression Line\n',fontsize=18)
plt.xlabel('\nFeature',fontsize=18)
plt.ylabel('Target\n',fontsize=18)
plt.legend()
```

### Evaluation


```python
#a) r2score (tells the relationship bw original values and predicted values,range 0 to 1)
r2=r2_score(Y,y_pred)    #1st parameter is original values & 2nd is predicted values
print("r2 score is: ",r2)

#mean score
mse=mean_squared_error(Y,y_pred)   #mse(error value)->squared difference bw original and predicted value.o to +infinity.close to 0 is better.
print("mse is: ",mse)
```
